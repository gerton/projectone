```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import pandas as pd

## generate random uniform points on a unit circle
def random_circle(N):
    nn = int( 1.28*N + 6*np.sqrt(0.4*N) )   ##  add six standard deviations so we have enough points for subsampling
    sample = ((2 * np.random.sample( 2 * nn ) - 1)[:2*N]).reshape( (N,2) )
    x, y = sample[:,0], sample[:,1]
    return sample[ x*x + y*y <= 1, : ]
```

```python
N = 2000000
R = 1.5
X = random_circle(N)
angle = np.random.sample(N) * 2 * 3.1415926
rim = np.array( (R*np.sin(angle), R*np.cos(angle)) ).T
df1 = pd.DataFrame( data=X, columns=["x","y"] )
df2 = pd.DataFrame( data=rim, columns=["x","y"])
sns.histplot(data = df1, x="x", y="y", bins=(100,100), cbar=True, cmap="RdYlGn_r")
sns.histplot(data = df2, x="x", y="y", bins=(100,100), cbar=True)
plt.show()
```
